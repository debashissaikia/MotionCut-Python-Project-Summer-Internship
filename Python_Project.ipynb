**Python project 1 :  A Basic Quiz Game**

```python
import time
print("Hello !!! ")
time.sleep(2)
name = input("Hey there !!! May I know your name please ?:")
print( name,", Welcome to the pc quiz game")
time.sleep(2)
print(name,", Do you want play the game ? " )
playing = input("Yes/no:").upper()
if playing != "YES" :
  print("Ohh .....")
  quit()
else:
  print("------------------------------------")
  print("----------Horrayyyy!!!!-------------")
  print("------------------------------------")
  time.sleep(2)
  print('''---------Instructions---------------
          Questions will be displayed one by one .
          There will be four options in each questions.
          You just need to type the correct option only .
          (Just write the option no. A, B, C or D no need write the answer) ''')
  print("------------------------------------")
  print("-----Best of Luck!!!!!--------------")

  time.sleep(3)


  score = 0

  Q1 = input('''>> Which of the following is the tallest statue in the world :
              A. Statue of Liberty
              B. Statue of Unity
              C. Statue of Belief
              D. Spring Temple of Buddha''').upper()
  if Q1 == "B":
    print("CORRECT!!!")
    score += 1
  else:
    print("INCORRECT")
  print("Statue of Unity situated in Gujarat,India is the tallest statue in the world")

  Q2 = input('''>> What is the most abundant gas in the earth's atmosphere :
              A. Oxygen
              B. Helium
              C. Hydrogen
              D. Nitrogen''').upper()
  if Q2 == "D":
    print("CORRECT!!!")
    score += 1
  else:
    print("INCORRECT")
  print("Nitrogen is the most abundant gas in the earth's atmosphere with more than 70% of its contribution to the atmosphere")

  Q3 = input('''>> Which of the following elements was discovered in space  :
              A. Hydrogen
              B. Helium
              C. Iron
              D. Aluminium''').upper()
  if Q3 == "B":
    print("CORRECT!!!")
    score += 1
  else:
    print("INCORRECT")
  print("Helium is a very rare gas in the earth's atmosphere. ")

  Q4 = input('''>> The chemical formula of gold is  :
              A. Au
              B. Ag
              C. Go
              D. Gd''').upper()
  if Q4 == "A":
    print("CORRECT!!!")
    score += 1
  else:
    print("INCORRECT")
  print("It is Au derived from the Latin name of gold Aurum")

  Q5 = input('''>> Who is considered as the Father of Genetics :
              A. Darwin
              B. Robert Hooke
              C. Gregor Mendel
              D. JBS Haldane''').upper()
  if Q5 == "C":
    print("CORRECT!!!")
    score += 1
  else:
    print("INCORRECT")
  print("Gregor Johann Mendel is considered as the Father of genetics for his famous set of equations with peas")
  time.sleep(2)

  Final_score = int(score/5 * 100)

  print("CONGRATUALATIONS !!!!! YOU HAVE COMPLETED THE QUIZ")
  time.sleep(2)
  print("Wait for a while , please .... Your score is under processing")
  print("                          ...                            ")
  time.sleep(6)
  print("Your score is :", Final_score,"%")

  if Final_score <= 40:
    print("GOOD, Try to make it better")
  elif Final_score <= 70:
    print("VERY GOOD")
  else:
    print("EXCELLENT")



```
